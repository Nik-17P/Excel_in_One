```python
import pandas as pd
import glob
import os

# Путь к папке с файлами
folder_path = "E:\Python_evening\эксель в один документ"

# Поиск всех файлов, начинающихся с "list" и заканчивающихся на ".xlsx"
file_paths = glob.glob(os.path.join(folder_path, "list*.xlsx"))

# Создаем пустой DataFrame для объединения данных
combined_data = pd.DataFrame()

# Читаем данные из каждого файла и добавляем их в общий DataFrame
for file in file_paths:
    try:
        # Чтение первого листа из файла
        data = pd.read_excel(file)
        combined_data = pd.concat([combined_data, data], ignore_index=True)
    except Exception as e:
        print(f"Ошибка при обработке файла {file}: {e}")

# Сохранение объединенных данных в новый Excel файл
output_file = os.path.join(folder_path, "combined_data.xlsx")
combined_data.to_excel(output_file, index=False)

print(f"Объединенные данные сохранены в файл: {output_file}")
```

```python
import pandas as pd
import plotly.express as px
from mapclassify import NaturalBreaks

# Путь к созданному файлу
file_path = "E:\Python_evening\эксель в один документ\combined_data_.xlsx"

# Чтение данных с листа "Лист2"
sheet_name = "Лист2"
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Извлечение данных
x_raw = data.iloc[:, 1]  # Второе поле
y_raw = data.iloc[:, 3]  # Четвертое поле
labels = data.iloc[:, 0]  # Названия предприятий (первое поле)

# Рассчитываем средние значения
x_mean = x_raw.mean()
y_mean = y_raw.mean()

# Преобразуем данные в отклонения от нормы
data["Отклонение X"] = x_raw# / x_mean
data["Отклонение Y"] = y_raw# / y_mean

# Разбиение на 10 классов методом естественных интервалов
nb = NaturalBreaks(data["Отклонение X"], k=10)
data["Класс X"] = nb.yb  # Классы (0, 1, ..., 9)

# Задание цветов от красного к синему
colorscale = px.colors.sequential.RdBu

# Создание интерактивной карты
fig = px.scatter(
    data,
    x="Отклонение X",
    y="Отклонение Y",
    text=labels,
    color="Класс X",
    color_continuous_scale=colorscale,
    title="Карта рыночных отношений",
    labels={"Отклонение X": "Отклонение по X", "Отклонение Y": "Отклонение по Y", "Класс X": "Класс"},
    template="plotly",
)

# Настройка отображения текста
fig.update_traces(textposition="top right", marker=dict(size=10, opacity=0.7))

# Добавление линий для средних значений
fig.add_shape(
    type="line",
    x0=data["Отклонение X"].min(),
    x1=data["Отклонение X"].max(),
    y0=0,
    y1=0,
    line=dict(color="Black", width=1, dash="dash"),
)
fig.add_shape(
    type="line",
    x0=0,
    x1=0,
    y0=data["Отклонение Y"].min(),
    y1=data["Отклонение Y"].max(),
    line=dict(color="Black", width=1, dash="dash"),
)

# Изменение размера графика
fig.update_layout(
    width=1024,
    height=768,
)

# Добавление кнопок для управления подписями
fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="right",
            x=0.7,
            y=1.2,
            showactive=True,
            buttons=[
                dict(
                    label="Включить подписи",
                    method="update",
                    args=[{"texttemplate": "%{text}"}],
                ),
                dict(
                    label="Выключить подписи",
                    method="update",
                    args=[{"texttemplate": ""}],
                ),
            ],
        )
    ]
)

# Показ графика
fig.show()
```
